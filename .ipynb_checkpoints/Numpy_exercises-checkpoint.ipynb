40. Write a NumPy program to compute the x and y coordinates for points on a sine curve and plot the points using matplotlib

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(360)
x_gra=x/(2*np.pi)
```

```python
y=np.sin(x_gra)
```

```python
plt.plot(x_gra,y)
```

```python
x_gra=x*np.pi/180
y=np.sin(x_gra)
```

```python
plt.plot(x_gra,y)
```

42. Write a NumPy program to add elements in a matrix. If an element in the matrix is 0, we will not add the element below this element

```python
ar42=np.eye(5)
ar42
```

```python
[ro,co]=np.shape(ar42)
print(ro,co)
```

```python
result=np.zeros_like(ar42)
for i in range(ro):
    if i==0:
        for j in range(co):
            result[i,j]=ar42[i,j]+np.random.randn(1)
    elif i>=1:
        for j in range(co):
            if ar42[i-1,j]!=0:
                result[i,j]=ar42[i,j]+np.random.randn(1)
```

```python
result
```

47. Write a NumPy program to create a one dimensional array of forty pseudo-randomly generated values. Select random numbers from a uniform distribution between 0 and 1.

```python
ar47=np.zeros(40)
ar47
```

```python
for i in ar47:
    ar47[i]=ar47[i]+np.random.normal(1)
```

```python
ar47=np.random.rand(40)
ar47
```
